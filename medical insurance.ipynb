```python
# encoding
insu_data.replace({'sex':{'male':0,'female':1}},inplace=True)
```

```python
insu_data.replace({'region':{'southeast':0,'southwest':1,'northeast':2,'northwest':3}},inplace=True)
```

```python
# insu_data.head()
```

```python
insu_data.replace({'smoker':{'yes':0,'no':1}},inplace=True)
insu_data.head()
```

```python
X=insu_data.drop(columns=('charges'),axis=1)
Y=insu_data['charges']
```

```python
print(X)
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
regressor=LinearRegression()
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
regressor.fit(X_train,Y_train)

```

```python
prediction_train=regressor.predict(X_train)
```

```python
r2_train=metrics.r2_score(Y_train,prediction_train)

print('r squarred error is:',r2_train)
```

```python
prediction_test=regressor.predict(X_test)
r2_test=metrics.r2_score(Y_test,prediction_test)
print("r squarred eror is :",r2_test)
```
